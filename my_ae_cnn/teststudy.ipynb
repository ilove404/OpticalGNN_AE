```python
from matplotlib import pyplot as plt
import numpy as np
from numpy import pi as pi

def awgn(x, snr, seed=7):
    """
    加入高斯白噪声 Additive White Gaussian Noise
    :param x: 原始信号
    :param snr: 信噪比
    :return: 加入噪声后的信号
    """
    np.random.seed(seed)  # 设置随机种子
    snr = 10 ** (snr / 10.0)
    xpower = np.sum(x ** 2) / len(x)
    npower = xpower / snr
    noise = np.random.randn(len(x)) * np.sqrt(npower)
    return x + noise

nums=1600
k=1
wid=40
long=40
SNR = 5
source=np.arange(nums)*pi/nums*8
testdata1=np.cos(source+pi/4)
testdata2=np.cos(source+3*pi/4)
testdata3=np.cos(source+5*pi/4)
testdata4=np.cos(source+7*pi/4)
testdata1_noise=awgn(testdata1,SNR)
testdata2_noise=awgn(testdata2,SNR)
testdata3_noise=awgn(testdata3,SNR)
testdata4_noise=awgn(testdata4,SNR)
plt.plot(source,testdata1_noise)
plt.plot(source,testdata2_noise)
plt.plot(source,testdata3_noise)
plt.plot(source,testdata4_noise)
```
